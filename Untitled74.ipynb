```python
y_train = train.SalePrice.values
all_data = pd.concat((train, test)).reset_index(drop=True)
all_data.drop(['SalePrice'], axis=1, inplace=True)
```

```python
all_data.isnull().sum()[all_data.isnull().sum()>0]
```

```python
all_data=all_data.drop(['Alley','Fence','FireplaceQu','PoolQC','Utilities'],axis=1)
missing_col_NA = ["MasVnrType","GarageType","MiscFeature","GarageQual","GarageCond","GarageFinish"]
missing_col_0 = ["MasVnrArea","GarageCars","GarageArea"]
for col in missing_col_NA:
    all_data[col].fillna("NA",inplace=True)    
for col in missing_col_0:
    all_data[col].fillna(0,inplace=True)
```

```python
missing_col_mode = ["MSZoning","KitchenQual","Functional","SaleType","Electrical","Exterior1st","Exterior2nd"]
for col in missing_col_mode:
    all_data[col].fillna(all_data[col].mode()[0],inplace=True)
all_data['LotFrontage'] = all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x:x.fillna(x.median()))
```

```python
all_data.BsmtCond.fillna(all_data.BsmtCond.mode()[0],inplace=True)
all_data.BsmtQual.fillna(all_data.BsmtQual.mode()[0],inplace=True)
all_data.BsmtExposure.fillna(all_data.BsmtExposure.mode()[0],inplace=True)
all_data.BsmtFinSF1.fillna(0,inplace=True)
all_data.BsmtFinSF2.fillna(0,inplace=True)
all_data.BsmtFinType1.fillna("NA",inplace=True)
all_data.BsmtFinType2.fillna("NA",inplace=True)
all_data.GarageYrBlt.fillna("NA",inplace=True)
all_data.TotalBsmtSF .fillna(all_data.TotalBsmtSF .dropna().median(),inplace=True)
all_data.BsmtFullBath.fillna(all_data.BsmtFullBath.dropna().median(),inplace=True)
all_data.BsmtHalfBath.fillna(all_data.BsmtHalfBath.dropna().median(), inplace=True)
all_data.BsmtUnfSF.fillna(all_data.BsmtHalfBath.dropna().median(), inplace=True)
NoBmstIndex = (pd.isnull(all_data["BsmtCond"])==True)&(pd.isnull(all_data["BsmtQual"])==True)&(pd.isnull(all_data["BsmtQual"])==True)
all_data.loc[NoBmstIndex,"BsmtCond"] =all_data.loc[NoBmstIndex,"BsmtCond"].fillna('NA')
all_data.loc[NoBmstIndex,"BsmtQual"] =all_data.loc[NoBmstIndex,"BsmtQual"].fillna("NA")
all_data.loc[NoBmstIndex,"BsmtExposure"] =all_data.loc[NoBmstIndex,"BsmtExposure"].fillna("NA")
```

```python
all_data.isnull().sum()[all_data.isnull().sum()>0]
```

```python
numeric_feats = all_data.dtypes[all_data.dtypes != "object"].index

# Check the skew of all numerical features
skewed_feats = all_data[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(10)
```

```python
skewness = skewness[abs(skewness) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    #all_data[feat] += 1
    all_data[feat] = boxcox1p(all_data[feat], lam)
```

```python
all_data = pd.get_dummies(all_data)
print(all_data.shape)
```

```python
x_train = all_data[:1458]
x_test = all_data[1458:]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,Y_train,Y_valid = train_test_split(x_train,y_train,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbdt = GradientBoostingRegressor(alpha=0.5,n_estimators=300,max_depth=2)
gbdt.fit(X_train,Y_train)
```

```python
predictions1 = gbdt.predict(X_valid)
from sklearn.metrics import mean_squared_error
print ("RMSE  is : \n",mean_squared_error(Y_valid,predictions1))
```

```python
from sklearn.model_selection import cross_val_score
def rmse_cv(model,X_train,y):
    rmse = np.sqrt(-cross_val_score(model, X_train, y, scoring="neg_mean_squared_error",cv = 5))
    return rmse
```

```python
gbdt_rmse = rmse_cv(gbdt,X_train,Y_train)
print ("{:.5f}: +/- {:.5f}".format(gbdt_rmse.mean(),gbdt_rmse.std()))
```
